```python
import pandas as pd
import requests
```

```python
def get_info(keggpathwayid):
    url = f"https://biodbnet-abcc.ncifcrf.gov/webServices/rest.php/biodbnetRestApi.json?method=db2db&input=keggpathwayid&inputValues={keggpathwayid}&outputs=genesymbol"
    response = requests.get(url,timeout = 600)
    if response.status_code == 200:
        data = response.json()
        gene_symbols = data["0"]["outputs"]["Gene Symbol"]
        return gene_symbols
    else:
        return None
```

```python
df = pd.read_excel('./KEGG人类所有通路 - 副本.xlsx')
ids = df['KEGGID'].to_list()
```

```python
error_list = {}
all_list = {}
# df = pd.DataFrame()
for id in ids:
    try:
        gene_symbols = get_info(id)
        if gene_symbols:
            all_list[id] =  pd.Series(gene_symbols)
    except Exception as e:
        error_list[id] = e
error_list
```

```python
all_list
```

```python
df = pd.DataFrame(all_list)
df
```

```python
df.to_csv('./db2db.csv',encoding='utf-8',index=False)
```
